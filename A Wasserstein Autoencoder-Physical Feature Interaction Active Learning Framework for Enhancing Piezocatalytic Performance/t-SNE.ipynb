```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import pandas as pd
import numpy as np
import os


def tsne_visualization(data, output_excel='tsne_results.xlsx', perplexity=30, learning_rate=200, n_iter=1000):
   
    
    if isinstance(data, pd.DataFrame):
        data = data.values  

    tsne = TSNE(n_components=2, perplexity=perplexity, learning_rate=learning_rate, n_iter=n_iter, random_state=42)
    tsne_result = tsne.fit_transform(data)

    
    tsne_df = pd.DataFrame(tsne_result, columns=['t-SNE component 1', 't-SNE component 2'])
    
    
    tsne_df.to_excel(output_excel, index=False)
    print(f"t-SNE results saved as {output_excel}")

   
    plt.figure(figsize=(8, 6))
    plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c='blue', marker='o')
    
    plt.title('t-SNE visualization')
    plt.xlabel('t-SNE component 1')
    plt.ylabel('t-SNE component 2')
    
    
    plt.show()


data_folder = 'reconstructeddata'


data_file = os.path.join(data_folder, 'reconstructed_x_data.csv')
data1_file = os.path.join(data_folder, '1.csv')


data = pd.read_csv(data_file)
data1 = pd.read_csv(data1_file)


tsne_visualization(data, output_excel='reconstructed_x_data_tsne.xlsx')
tsne_visualization(data1, output_excel='data1_tsne.xlsx')
```
